```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define campaigns and platforms
campaigns = [
    "Summer Sale", "Winter Promo", "Black Friday", "Holiday Deals",
    "New Year Sale", "Spring Campaign", "Flash Sale", "Cyber Monday",
    "Valentine Promo", "Easter Campaign"
]
platforms = ["Instagram", "Facebook", "Google Ads"]

# Generate 30 days of dates
start_date = datetime(2025, 11, 1)
dates = [start_date + timedelta(days=i) for i in range(30)]

# Generate dataset
rows = []
for date in dates:
    for campaign in campaigns:
        platform = np.random.choice(platforms)
        impressions = np.random.randint(5000, 50000)
        clicks = np.random.randint(100, impressions//5)  # clicks <= 20% of impressions
        conversions = np.random.randint(10, clicks//2 + 1)  # conversions <= 50% of clicks
        spend = round(np.random.uniform(100, 1000), 2)
        rows.append([campaign, platform, date.strftime("%Y-%m-%d"), impressions, clicks, conversions, spend])

# Create DataFrame
df = pd.DataFrame(rows, columns=["Campaign", "Platform", "Date", "Impressions", "Clicks", "Conversions", "Spend"])

# Save to CSV
df.to_csv("campaign_data.csv", index=False)

print("Dataset generated: campaign_data.csv")
```
